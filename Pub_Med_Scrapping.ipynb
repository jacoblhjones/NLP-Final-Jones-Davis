```python



def scrape_pdf_links_with_selenium(url):
    options = Options()
    options.add_argument('--headless')  # Run Chrome in headless mode (no GUI)
    options.add_argument('--no-sandbox')  # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resources problem
    driver = gs.Chrome()

    try:
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        pdf_links = []
        for link in soup.find_all('a', href=True):
            href = link['href']
            if href.lower().endswith('.pdf'):
                if not href.startswith('http'):
                    href = requests.compat.urljoin(url, href)
                pdf_links.append(href)

        return pdf_links
    finally:
        driver.quit()

# Example usage
url = 'https://www.mdcalc.com/calc/10008/revised-schwartz-equation-glomerular-filtration-rate-gfr-2009#evidence'
pdf_links = scrape_pdf_links_with_selenium(url)
for link in pdf_links:
    print(link)


```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests
import google_colab_selenium as gs
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

def scrape_gov_org_links_with_selenium(url):
    options = Options()
    options.add_argument('--headless')  # Run Chrome in headless mode (no GUI)
    options.add_argument('--no-sandbox')  # Bypass OS security model
    options.add_argument('--disable-dev-shm-usage')  # Overcome limited resources problem
    driver = gs.Chrome()

    try:
        driver.get(url)
        print(f"Opened URL: {url}")

        # Wait for the page to fully load
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )

        # Parse the page content
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        gov_org_links = []
        for link in soup.find_all('a', href=True):
            href = link['href'].strip()  # Remove any leading/trailing whitespace
            if '.gov' in href or '.org' in href:
                if not href.startswith('http'):
                    href = requests.compat.urljoin(url, href)
                gov_org_links.append(href)

        return gov_org_links

    except Exception as e:
        print(f"Error occurred: {str(e)}")
        return []

    finally:
        driver.quit()

# Example usage
url = 'https://www.mdcalc.com/calc/10008/revised-schwartz-equation-glomerular-filtration-rate-gfr-2009#evidence'
gov_org_links = scrape_gov_org_links_with_selenium(url)
if gov_org_links:
    for link in gov_org_links:
        print(link)
else:
    print("No .gov or .org links found or an error occurred during scraping.")


```

```python
from urllib.parse import urlparse

def group_links_by_domain(data):
    """
    Groups links by their domain names.

    Parameters:
        data (dict): A dictionary where keys are IDs and values are lists of links.

    Returns:
        dict: A dictionary where keys are domain names and values are lists of links corresponding to those domains.
    """
    domain_dict = {}
    for links in data.values():
        for link in links:
            domain = urlparse(link).netloc
            if domain not in domain_dict:
                domain_dict[domain] = []
            domain_dict[domain].append(link)
    return domain_dict

# Example usage:

domains = group_links_by_domain(data_dict)

```

```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup
from tqdm import tqdm
import nest_asyncio

# Apply the nest_asyncio patch
nest_asyncio.apply()

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

async def fetch(url, session):
    try:
        async with session.get(url) as response:
            response.raise_for_status()
            text = await response.text()
            return text
    except Exception as e:
        # print(f"Error fetching {url}: {e}")
        return None

async def get_doi_from_url(url):
    async with aiohttp.ClientSession(headers=headers) as session:
        html = await fetch(url, session)
        if html:
            try:
                soup = BeautifulSoup(html, 'html.parser')
                doi_element = soup.find('a', {'data-ga-action': 'DOI'})
                if doi_element:
                    return doi_element.get('href')
                return None
            except Exception as e:
                # print(f"Error parsing the content from {url}: {e}")
                return None
        return None

async def get_doi_from_pubmed(pubmed_dict):
    results = {}
    for calc_id, urls in tqdm(pubmed_dict.items(), desc="Processing calc_ids"):
        # print(f"\nProcessing calc_id: {calc_id}")
        tasks = [get_doi_from_url(url) for url in urls]
        dois = await asyncio.gather(*tasks)
        results[calc_id] = dois
        # print(f"DOIs for calc_id {calc_id}: {dois}")
    return results


```

```python


import asyncio
import aiohttp
from bs4 import BeautifulSoup
from tqdm import tqdm
import nest_asyncio

# Apply the nest_asyncio patch
nest_asyncio.apply()

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 '
                  '(KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

async def fetch(url, session):
    try:
        async with session.get(url) as response:
            response.raise_for_status()
            text = await response.text()
            return text
    except Exception as e:
        # print(f"Error fetching {url}: {e}")
        return None

async def get_abstract_from_url(url):
    async with aiohttp.ClientSession(headers=headers) as session:
        html = await fetch(url, session)
        if html:
            try:
                soup = BeautifulSoup(html, 'html.parser')
                abstract_div = soup.find('div', class_='abstract-content selected')
                if abstract_div:
                    return abstract_div.get_text(strip=True)
                return None
            except Exception as e:
                # print(f"Error parsing the content from {url}: {e}")
                return None
        return None

async def get_abstracts_from_pubmed(pubmed_dict):
    results = {}
    for calc_id, urls in tqdm(pubmed_dict.items(), desc="Processing calc_ids"):
        # print(f"\nProcessing calc_id: {calc_id}")
        tasks = [get_abstract_from_url(url) for url in urls]
        abstracts = await asyncio.gather(*tasks)
        results[calc_id] = abstracts
        # print(f"Abstracts for calc_id {calc_id}: {abstracts}")
    return results


# Run the async function
async def main():
    abstract_dict = await get_abstracts_from_pubmed(data_dict)
    return abstract_dict

abstract_dict = asyncio.run(main())
```

```python
def count_pdf_links(link_dict):
    count = 0
    for key, link_list in link_dict.items():
        if link_list is not None:
            for link in link_list:
                if link and ('pdf' in link.lower() or 'full' in link.lower()):
                    count += 1
    return count



def find_keys_with_empty_lists(dictionary):
    empty_list_keys = []
    for key, value in dictionary.items():
        if isinstance(value, list) and len(value) == 0:
            empty_list_keys.append(key)
    return len(empty_list_keys), empty_list_keys
```

```python
def count_non_none_first_elements(input_dict):
    count = 0
    for key, value in input_dict.items():
        if len(value) > 0 and value[0] is not None:
            count += 1
    return count

def count_non_none_elements(input_dict):
    count = 0
    for key, value in input_dict.items():
        count += sum(1 for element in value if element is not None)
    return count

```

```python
def replace_none_values(primary_dict, secondary_dict):
    # Iterate through keys in primary_dict
    for key in primary_dict:
        # Check if key exists in secondary_dict and its value is a list
        if key in secondary_dict and isinstance(secondary_dict[key], list):
            # Iterate through the list in secondary_dict
            for i, value in enumerate(secondary_dict[key]):
                # Replace None values with corresponding index values from primary_dict
                if value is None:
                    secondary_dict[key][i] = primary_dict[key][i]
```

```python
def apply_function_to_lists(input_dict, func):
    # Create a new dictionary to store transformed lists
    transformed_dict = {}

    # Iterate through each key-value pair in the input dictionary
    for key, lst in input_dict.items():
        # Apply the function to each element in the list using a list comprehension
        transformed_list = [func(item) for item in lst]
        # Store the transformed list in the new dictionary
        transformed_dict[key] = transformed_list

    return transformed_dict
```
